```python
import pandas as pd
import numpy as np
```

```python

data = pd.read_csv(r'analysed_df.csv')

df = pd.read_csv(r'KAG_energydata_complete.csv')
```

```python
RH_out_column = df['RH_out']

# Add 'RH_out' column to data
data['RH_out'] = RH_out_column

```

```python

data
```

```python
# Group the data by the 'Cluster' column
grouped = data.groupby('Cluster')
```

```python
# Create a new DataFrame with the desired columns and rows
profiled_data = pd.DataFrame({
    'min_Appliances': grouped['Appliances'].min(),
    'avg_Appliances': grouped['Appliances'].mean(),
    'max_Appliances': grouped['Appliances'].max(),
    'min_PredictedEnergy': grouped['PredictedEnergy'].min(),
    'avg_PredictedEnergy': grouped['PredictedEnergy'].mean(),
    'max_PredictedEnergy': grouped['PredictedEnergy'].max(),
    'min_T_out': grouped['T_out'].min(),
    'avg_T_out': grouped['T_out'].mean(),
    'max_T_out': grouped['T_out'].max(),
    'min_RH_out': grouped['RH_out'].min(),
    'avg_RH_out': grouped['RH_out'].mean(),
    'max_RH_out': grouped['RH_out'].max(),
    'min_Tdewpoint': grouped['Tdewpoint'].min(),
    'avg_Tdewpoint': grouped['Tdewpoint'].mean(),
    'max_Tdewpoint': grouped['Tdewpoint'].max()
}, index=pd.Series(range(grouped.ngroups), name='Clusters'))
```

```python
# Transpose the DataFrame to make Cluster values as columns
profiled_data = profiled_data.T
```

```python
# Print the new DataFrame
profiled_data
```

```python
profiled_data.to_csv('profiling2.csv', index=True)
```
